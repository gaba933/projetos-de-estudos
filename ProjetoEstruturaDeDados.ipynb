Escreva um programa cliente que utilize o tipo abstrato de dados Pilha para simular uma sessão com um caixa de banco. Contrariamente à maioria dos bancos,este decidiu que o último cliente a chegar será sempre o primeiro a ser obtido.

Crie classes que representem informações sobre um cliente de um banco e uma transação. Para cada cliente você precisa armazenar um nome, saldo na conta corrente e uma referência à transação. Para cada transação, você precisa armazenar o tipo de transação (depósito ou retirada) e o valor da transação.

Após cada cinco clientes serem processados, exiba o tamanho da pilha e o nome do próximo cliente a ser atendido.

```python
# Criação das classes
class Cliente:
  def __init__(self,nome,saldo,conta):
    self.nome = nome
    self.saldo = saldo
    self.conta = conta

class Conta:
  def __init__(self,tipo,valor):
    self.tipo = tipo
    self.valor = valor

class Pilha:
  def __init__(self):
    # Inicializa a pilha com uma lista vazia:
    self.items= []

  def empty(self):
    # Identifica se a pilha estiver vazia:
    return len(self.items) == 0
  
  def stack(self,item):
    # Acrescenta um elemento no topo da pilha:
    self.items.append(item)

  def unstack(self):
    # Remove um elemento no topo da pilha:
    if self.empty():
      return None
    return self.items.pop()

  def size(self):
    # Retorna o número de elementos na pilha:
    return len(self.items)

# Implementação da pilha:
pilha = Pilha()
clientes_processados = 0

print("|---| Simulação de Caixa de Banco |---|")

# Implementação do menu de opções:
while True:
  print("\n")
  print("1. Inserir manualmente os clientes.")
  print("2. Simular uma lista pré-determinada de clientes.")
  print("0. Sair.")
  opc = input("Selecione uma opção:")

  if opc == "1":
    while True:
      if clientes_processados % 5 == 0:
        print("Tamanho da fila do banco: ", pilha.size())
        if not pilha.empty:
          print("Próximo cliente a ser processado: ", pilha.unstack().nome)
      
      # Leitura dos dados do cliente:
      print("\n--- Cadastro de Cliente ---")
      nome = input("Nome do cliente (ou 'sair' para encerrar): ")
      if nome.lower() == "sair":
        break
      
      saldo = float(input("Saldo do cliente: "))

      while True:
        tipo_conta = input("Tipo de transação (depósito ou retirada): ")
        if tipo_conta == "depósito" or tipo_conta == "retirada":
          break
        else:
          print("Opção inválida. Digite 'depósito' ou 'retirada'.")

      valor_conta = float(input("Valor da transação: "))

      # Implementa o objeto "Conta":
      conta = Conta(tipo_conta,valor_conta)
      # Implementa o objeto "Cliente":
      cliente = Cliente(nome,saldo,conta)
      # Insere o elemento na pilha:
      pilha.stack(cliente)
      # Incrementa o contador de clientes processados:
      clientes_processados += 1
  
  elif opc == "2":
    lista_clientes = [
        # Exemplo de lista de clientes:
        Cliente("Scott Richards",100.0,Conta("depósito",500.0)),
        Cliente("Riley Gordon",5000.0,Conta("retirada", 1000.0)),
        Cliente("Ryan Leach",200.0,Conta("retirada", 50.0)),
        Cliente("Amari Nicholson",3000.0,Conta("depósito", 2000.0)),
        Cliente("Marcel Odom",1500.0,Conta("retirada",300.0)),
        Cliente("Isabella Bailey",400.0,Conta("depósito",100.0)),
        Cliente("Nicole Fox",1000.0,Conta("retirada",2000.0)),
        Cliente("Raylee Green",2500.0,Conta("depósito",1500.0)),
        Cliente("Chris Powell",800.0,Conta("retirada",100.0)),
        Cliente("Brett Ellison",500.0,Conta("depósito",300.0))
    ]
    
    print("\nExecutando lista de clientes...")

    # Imprime os dados dos clientes:
    for i, cliente in enumerate(lista_clientes, start=1):
      pilha.stack(cliente)
      clientes_processados += 1

      if clientes_processados % 5 == 0:
        print("\nTamanho da pilha: ", pilha.size())
        if not pilha.empty():
          proximo_cliente = pilha.unstack()
          print("Próximo cliente a ser atendido: ", proximo_cliente.nome)
          pilha.stack(proximo_cliente)
          # Mantém o contador para o próximo ciclo:
          clientes_processados -= 1
  
  elif opc == "0":
    break
  
  else:
    print("Opção inválida.")
```

Crie uma função recursiva chamada ComparaArvores que receba duas árvores e compare se são iguais.

As árvores são iguais caso elas tenham exatamente a mesma estrutura e os mesmos valores armazenados exatamente nos mesmos nós. 

```python
class Node:
  def __init__(self,data):
    self.data = data
    self.left = None
    self.right = None

  def comparaArvore(arvoreX,arvoreY):
    # Caso padrão 1: se as árvores estiverem vazias:
    if arvoreX is None and arvoreY is None:
      return True

    # Caso padrão 2: se uma das árvores estiver vazia:
    if arvoreX is None or arvoreY is None:
      return False

    # Caso padrão 3: se as árvores forem diferentes:
    if arvoreX.data != arvoreY.data:
      return False
    
    # Compara as subárvores esquerdas e direitas recursivamente.
    return (Node.comparaArvore(arvoreX.left,arvoreY.left) and Node.comparaArvore(arvoreX.right,arvoreY.right))

# -----------------------------------------------------------------------------
# 1º Verificação
print("|--------------------------------|")
print("|-------| 1º Verificação |-------|")
arvoreX = Node(1)
arvoreX.left = Node(2)
arvoreX.right = Node(3)

arvoreY = Node(1)
arvoreY.left = Node(2)
arvoreY.right = Node(3)

print("Conclusão de verificação: ",Node.comparaArvore(arvoreX,arvoreY)) # Retorno: verdadeiro
# -----------------------------------------------------------------------------
# 2º Verificação
print("|--------------------------------|")
print("|-------| 2º Verificação |-------|")

arvoreX = Node(1)
arvoreX.left = Node(2)
arvoreX.left = Node(4)
arvoreX.right = Node(3)
arvoreX.right = Node(5)

arvoreY = Node(1)
arvoreY.left = Node(2)
arvoreY.left = Node(4)
arvoreY.right = Node(3)
arvoreY.right = Node(5)

print("Conclusão de verificação: ",Node.comparaArvore(arvoreX,arvoreY)) # Retorno: verdadeiro
print("|--------------------------------|")
# -----------------------------------------------------------------------------
# 3º Verificação
print("|-------| 3º Verificação |-------|")
arvoreX = Node(0)

arvoreY = Node(0)

print("Conclusão de verificação: ",Node.comparaArvore(arvoreX,arvoreY)) # Retorno: verdadeiro
print("|--------------------------------|")
# -----------------------------------------------------------------------------
# 4º Verificação
print("|-------| 4º Verificação |-------|")
arvoreX = Node(2)
arvoreX.left = Node(4)
arvoreX.right = Node(6)

arvoreY = Node(1)
arvoreY.left = Node(3)
arvoreY.right = Node(5)

print("Conclusão de verificação: ",Node.comparaArvore(arvoreX,arvoreY)) # Retorno: falso
print("|--------------------------------|")
# -----------------------------------------------------------------------------
# 5º Verificação
print("|-------| 5º Verificação |-------|")
arvoreX = Node(1)
arvoreX.left = Node(2)
arvoreX.left = Node(4)
arvoreX.right = Node(3)
arvoreX.right = Node(5)

arvoreY = Node(1)
arvoreY.left = Node(3)
arvoreY.right = Node(5)

print("Conclusão de verificação: ",Node.comparaArvore(arvoreX,arvoreY)) # Retorno: falso
print("|--------------------------------|")
# -----------------------------------------------------------------------------
```

Modifique a árvore binária de busca ensinada nesta disciplina e dê o nome de TreePath. Acrescente nessa árvore um método chamado caminhos. Esse método deve preencher um objeto da lista duplamente ligada fornecida junto com este notebook com a soma do valores dos nós que estão no caminho de cada um dos nós folhas até a raiz da árvore. Por exemplo, considere a árvore apresentada abaixo:

            ___________61___________
           |                        |
       ____34_____              ____67____
      |           |            |          |    
      19__        50__      __66        __94__
          |           |    |           |      |  
         20           55   64          70     95

Se a árvore acima fosse apresentada como entrada, o método caminhos deveria retornar uma lista duplamente ligada com os seguintes valores: [134, 200, 258, 292, 317], que é formada pelas seguites somas: [61+34+19+20, 61+34+50+55, 61+67+66+64, 61+67+94+70, 61+67+94+95].

```python
# Define a classe Node que representa um nó em uma árvore binária.
class Node:
  def __init__(self,data):
    self.data = data
    self.right = None
    self.left = None

# Define a classe TreePath que representa uma árvore binária
class TreePath:
  def __init__(self):
    self.root = None
  
  def insert(self,data,node=None):
    # Insere um novo nó na árvore com o valor do elemento desejado.
    # Se a raiz não existe, cria um novo nó como raiz.
    # Caso contrário, percorre a árvore para encontrar a posição correta para inserir o elemento desejado.
    if self.root is None:
      self.root = Node(data)
    else:
      if node is None:
        node = self.root
      if data < node.data:
        if node.left is None:
          node.left = Node(data)
        else:
          self.insert(data,node.left)
      else:
        if node.right is None:
          node.right = Node(data)
        else:
          self.insert(data,node.right)

  def strInorder(self,node=-1,info=''):
    # Retorna uma representação em ordem dos elementos da árvore.
    # Se a raiz não existe, retorna uma string vazia.
    # Caso contrário, percorre a árvore em ordem recursivamente.
    if self.root is None:
      return ''
    else:
      if node == -1:
        node = self.root
      if node.data is not None:
        if node.left is not None:
          info += self.strInorder(node.left)
        info += '' + str(node.data)
        if node.right is not None:
          info += self.strInorder(node.right)
        return info
      else:
        return info
      
  def strPreorder(self,node=-1,info=''):
    # Retorna uma representação em pré-ordem dos elementos da árvore.
    # Se a raiz não existe, retorna uma string vazia.
    # Caso contrário, perecorre a árvore em pré-ordem recursivamente. 
    if self.root is None:
      return ''
    else:
      if node == -1:
        node = self.root
      if node.data is not None:
        info += '' + str(node.data)
        if node.left is not None:
          info += self.strPreorder(node.left)
        if node.right is not None:
          info += self.strPreorder(node.right)
        return info
      else:
        return info

  def strPostorder(self,node=-1,info=''):
    # Retorna uma representação em pós-ordem dos elementos da árvore.
    # Se a raiz não existe, retorna uma string vazia.
    # Caso contrário, perecorre a árvore em pós-ordem recursivamente. 
    if self.root is None:
      return ''
    else:
      if node == -1:
        node = self.root
      if node.data is not None:
        if node.left is not None:
          info += self.strPostorder(node.left)
        if node.right is not None:
          info += self.strPostorder(node.right)
        info +='' + str(node.data)
        return info
      else:
        return info
  
  def search(self,valor,anode=-1):
    # Realiza uma busca pelo elemento desejado da árvore.
    # Se o nó especificado por nulo (-1), inicia a busca pela raiz da árvore.
    if anode == -1:
      anode = self.root
    # Se o nó atual for nulo (None), retorna falso, indicando que o elemento não foi encontrado.
    if anode is None:
      return False
    # Se o nó atual for igual ao elemento desejado, retorna verdadeiro, indicando que o elemento foi encontrado.
    elif valor == anode.data:
      return True
    # Se o elemento desejado for menor que o nó atual, faz uma busca recursiva na subárvore esquerda.
    elif valor < anode.data:
      return self.search(valor,anode.left)
    # Caso contrário, faz uma busca recursiva na subárvore direita.
    else:
      return self.search(valor,anode.right)

  def path(self,node=None,sum=0,path_list=None):
    # Retorna uma lsita contendo as somas de todos os caminhos da raiz até as folhas das árvores
    # Se o nó especificado por nulo (None), inicia o caminho pela raiz da árvore.
    if node is None:
      node = self.root
    # Se a lsita de caminhos não for especificada, cria uma nova lista vazia.
    if path_list is None:
      path_list = []
    # Se o nó atual não for nulo (None), incrementa a soma com o valor do nó atual.
    if node is not None:
      sum += node.data
      # Se o nó atual for uma folha, adiciona a soma à lista de caminhos
      if node.left is None and node.right is None:
        path_list.append(sum)
      # Faz chamadas recursivas para os nós filhos, se existirem.
      if node.left:
        self.path(node.left,sum,path_list)
      if node.right:
        self.path(node.right,sum,path_list)
    # Retorna a lista de caminhos atualizada.
    return path_list

# Cria uma nova instância da classe TreePath.
tree = TreePath()

# Insere os valores na árvore.
tree.insert(61)
tree.insert(34)
tree.insert(67)
tree.insert(19)
tree.insert(50)
tree.insert(66)
tree.insert(94)
tree.insert(20)
tree.insert(55)
tree.insert(64)
tree.insert(70)
tree.insert(95)

# Calcula a lista de caminhos na árvore e armazena em path_list.
path_list = tree.path()

print(path_list) # [134, 200, 258, 292, 317]
```
